## Завдання 1


$50x^7 + 717x^6 + 675x^5 - 887x^4 - 791x^3 + 165x^2 + 96x -7 = 0$



### відокремлення коренів

За основною теоремою алгебри, враховуючи кратність, рівняння має 7 коренів. За теоремою 3 маємо формулу:

$$\cfrac{1}{1 + \frac{B}{|a_0|}} \leq |x_{*i}| \leq 1 + \frac{A}{|a_n|}, $$

$де \quad \,\, A = \max\{|a_6|,...,|a_0|\} = |a_4| = 887$

$\quad \quad \, B = \max\{|a_7|,...,|a_1|\} = |a_4| = 887$

тому:

$\quad \quad \, \cfrac{1}{1 + \frac{887}{7}} \leq |x_{*i}| \leq 1 + \frac{887}{50}
\quad \Rightarrow{} \quad 0.007829 \leq |x_{*i}| \leq 18.74$ 

додатні корені:

$\quad \quad 0.007829 \leq x_{*i}^+ \leq 18.74 $

від'ємні корені:

$\quad \quad -18.74 \leq x_{*i}^- \leq -0.007829 $

$\\~$

Розіб'ємо проміжки існування коренів на ще менші, і знайдемо ті, на яких функція змінює знак

```python
def func(x):
    return (50*x**7 + 717*x**6 + 675*x**5 - 887*x**4 - 791*x**3 + 165*x**2 + 96*x -7 )

def main():
    
    t =-18.74
    
    #Від'ємні корені
    while t < -0.007829:
        
        print(f"f({round(t,6)}) має знак ", end = '')
        
        if(func(t) > 0):
            print("+")
            
        else:
            print("-")
            
        t = t + 0.1
        
    print("\n\n--------------------------------------\n\n")
    
    t = 0.007829
    #додатні корені
    while t < 18.74:
        
        print(f"f({round(t,6)}) має знак ", end = '')
        
        if(func(t) > 0):
            print("+")
            
        else:
            print("-")
            
        t = t + 0.1
        
    
    
if __name__ == "__main__":
    main()
```

Отримали 7 проміжків, на яких функція змінює знак. Оскільки рівняння всього має 7 коренів, то на
кожному з цих проміжків буде лише один корінь

$(−13.24; −13.14) \quad (−1.34; −1.24) \quad (−0.94; −0.84) \quad (−0.44; −0.34) \quad (0.007829; 0.107829) \quad
(0.307829; 0.407829) \quad (0.907829; 1.007829)$

### метод бісекцій

```python
import math

bisection_1 = []

intervals = [(-13.24, -13.14), (-1.34, -1.24), (-0.94, -0.84), (-0.44, -0.34),
             (0.007829, 0.107829), (0.307829, 0.407829), (0.907829, 1.007829)]

eps = 0.00001

iterations = []

def print_(c,f):
    
    if c < 0:
        if func(c) < 0: print(f"{c} \t {func(c)}")
        else: print(f"{c} \t  {f(c)}")
    else:
        if func(c) < 0: print(f" {c} \t {func(c)}")
        else: print(f" {c} \t  {f(c)}")


print("       змінна                     функція")
 
iter = 0

for interval in intervals:
    
    a = interval[0]
    b = interval[1]
    c = (a+b)/2
    
    while b - a >= eps or abs(func(c)) >= eps:
        c = (a+b)/2
        
        if func(c) == 0:
            break
            
        if func(a)*func(c) < 0:
            b = c
        else:
            a = c
        
        iter = iter + 1
        
    bisection_1.append(c)
    print_(c,func)
    
iterations.append(iter)
print("\n\nкількість ітерацій: ", iter)
```

### метод хорд

спочатку відділимо проміжки ізоляції кореня, на яких друга похідна функції $f(x)$ зберігає знак.

$f(x) = 50x^7 + 717x^6 + 675x^5 - 887x^4 - 791x^3 + 165x^2 + 96x -7$

$f'(x) = 350 x^6 + 4302 x^5 + 3375 x^4 - 3548 x^3 - 2373 x^2 + 330 x + 96 $

$f''(x) = 2100 x^5 + 21510 x^4 + 13500 x^3 - 10644 x^2 - 4746 x + 330 $

```python
def second_derr(x):
    return (2100*x**5 + 21510*x**4 + 13500*x**3 - 10644*x**2 - 4746*x + 330)

intervals = [(-13.24, -13.14), (-1.34, -1.24), (-0.94, -0.84), (-0.44, -0.34),
             (0.007829, 0.107829), (0.307829, 0.407829), (0.907829, 1.007829)]

for interval in intervals:
    
    t = interval[0]
    
    while t < interval[1]:
        
        print(f"f({round(t,6)}) має знак ", end ='')
        
        if func(t) < 0:
            print("- ", end = '\t')
        elif func(t) == 0:
            print("0 ", end = '\t')
        else:
            print("+ ", end = '\t')
        
        print(f"f\'\'({round(t,6)}) має знак ", end ='')
        
        if second_derr(t) < 0:
            print("-")
        elif second_derr(t) == 0:
            print("0")
        else:
            print("+")
        
        
        t = t + 0.01
        
    print("----------------------------")
```

Як бачимо проміжок $(0.057829; 0.067829)$ потрібно дослідити більш детально

```python
# (0.057829; 0.067829)

t = 0.057829

while t <= 0.067829:
    
    print(f"f({round(t,6)}) має знак ", end ='')
        
    if func(t) < 0:
        print("- ", end = '\t') 
    elif func(t) == 0:
        print("0 ", end = '\t')
    else:
        print("+ ", end = '\t')
        
    print(f"f\'\'({round(t,6)}) має знак ", end ='')
        
    if second_derr(t) < 0:
        print("-")
    elif second_derr(t) == 0:
        print("0")
    else:
        print("+")
            
    t = t + 0.001
    
```

отже отримали сім проміжків, на яких знаходяться корені, і на яких друга похідна не змінює знак:
    
$(-13.23; -13.22), \quad (-1.31; -1.3), \quad (-0.88; -0.87), \quad (-0.37; -0.36), \quad (0.066829; 0.067829), \quad(0.367829;0.377829)$

$(0.987829,0.997829)$

тепер застосуємо метод хорд для цих проміжків

```python
new_intervals = [(-13.23, -13.22), (-1.31, -1.3),  (-0.88, -0.87),
                 (-0.37, -0.36),  (0.066829, 0.067829), (0.367829, 0.377829),
                 (0.987829,0.997829)]

secant_1 = []

eps = 0.00001

print("       змінна                     функція")

iter = 0

for interval in new_intervals:
    
    a = interval[0]
    b = interval[1]
    
    if second_derr(a)*func(a) > 0:
        N = a
        x_k1 = b 
    else:
        N = b
        x_k1 = a
        
    x_k2 = x_k1 + 0.00002 
    
    k = 0
    
    while abs(x_k2 - x_k1) >= eps or abs(func(x_k2)) >= eps:
        
        if k!= 0 : 
            x_k1 = x_k2
        
        x_k2 = x_k1 - (func(x_k1)*(x_k1 - N))/(func(x_k1) - func(N))
        
        k = 1
        
        iter = iter + 1
        
    secant_1.append(x_k2)    
    print_(x_k2, func) 
    

iterations.append(iter)
print("\n\nкількість ітерацій: ", iter)
                                               
```

### метод Ньютона


спочатку виберемо для кожного проміжка початкове наближення так, щоб виконувалась умова "$f(x) \cdot f''(x) > 0$"

```python
X = []
k = 0

intervals = [(-13.24, -13.14), (-1.34, -1.24), (-0.94, -0.84), (-0.44, -0.34),
             (0.007829, 0.107829), (0.307829, 0.407829), (0.907829, 1.007829)]


print("    проміжок          п. наближення")

for interval in intervals:
    
    if func(interval[0])*second_derr(interval[0]) > 0:
        X.append(interval[0])
    else:
        X.append(interval[1])
        
    print(f"({interval[0]};{interval[1]})   \t {X[k]}")
    
    k = k + 1
    
```

тепер для кожного проміжка використаймо метод Ньютона

```python
intervals = [(-13.24, -13.14), (-1.34, -1.24), (-0.94, -0.84), (-0.44, -0.34),
             (0.007829, 0.107829), (0.307829, 0.407829), (0.907829, 1.007829)]

newton_1 = []

iter = 0
k = 0
eps = 0.00001
step = 0

def first_derr(x):
    return (350*x**6 + 4302*x**5 + 3375*x**4 - 3548*x**3 - 2373*x**2 + 330*x + 96 )

for interval in intervals:
    
    x_k0 = X[k]
    x_k1 = x_k0 + 2*eps
    
    while abs(x_k1 - x_k0) >= eps or abs(func(x_k1)) >= eps:
        
        if step != 0:
            x_k0 = x_k1
        x_k1 = x_k0 - (func(x_k0)/first_derr(x_k0))
        
        step = 1
        iter = iter + 1
    
    step = 0
    k = k + 1
    print_(x_k1, func)
    newton_1.append(x_k1)
    
iterations.append(iter)
print(f"\nкількість ітерацій: {iter}")
        
```

## Завдання 2

tg$(0.4x+0.4) = x^2$

### Відокремлення коренів

Для Відокремлення коренів зобразимо $f(x)$ графічно

Як бачимо, існує безліч розв'язків рівняння $f(x) = 0$. Знайдемо ті, які розташовані на проміжку (-6;4). Але спочатку знайдемо вертикальні асимптоти $f(x)$, які належать проміжку (-6;4):

$0.4x + 0.4 = \cfrac{\pi}{2} + \pi n \quad \Rightarrow \quad x = \cfrac{\pi}{0.8} - 1 + \cfrac{\pi n}{0.4}$

1. $n = -1:  x = \cfrac{\pi}{0.8} - 1 - \cfrac{\pi}{0.4} = -4.92699$

2. $n = 0: x = \cfrac{\pi}{0.8} - 1 = 2.92699$

```python
def second_func(x):
    return(math.tan(0.4*x+0.4) - x**2)

t = -6

while t < 4:
    
    print(f"f({round(t,6)}) має знак ", end = '')

    
    if second_func(t) < 0:
        print("-")
    elif second_func(t) > 0:
        print("+")
    else:
        print("0")
        
    t = t + 0.1

    
```

Отримуємо 6 проміжків, на яких функція змінює знак, хоча на цьому проміжку лише 4 корені. Це пояснюється тим, що на проміжках (-5; -4.9) та (2.9; 3) знаходяться вертикальні асимптоти $x = -4.92699$ та $x = 2.92699$ відповідно, а, отже, на цих проміжках не буде коренів. Тому отримуємо 4 проміжки, на кожному яких знаходится лише один корінь: (-5.1; -5), (-0.5; -0.4), (1; 1.1),  (2.5;  2.6) 

### метод бісекцій

```python
intervals = [(-5.1,-5), (-0.5, -0.4), (1, 1.1),  (2.5,  2.6)]
iter = 0
eps = 0.00001

bisection_2 = []

print("       змінна                     функція")

for interval in intervals:
    
    a = interval[0]
    b = interval[1]
    c = (a+b)/2

    
    while b - a >= eps or abs(second_func(c)) >= eps:
        
        if second_func(c) == 0:
            break
        
        if second_func(a)*second_func(c) < 0:
            b = c
        elif second_func(b)*second_func(c) < 0:
            a = c
        
        c = (a+b)*0.5
        
        iter = iter + 1
        
    print_(c,second_func)
    bisection_2.append(c)
    
iterations.append(iter)
print(f"\nкількість ітерацій: {iter}")
```

### метод хорд

спочатку відділимо проміжки ізоляції кореня, на яких друга похідна функції $f(x)$ зберігає знак.

$f(x) =$ tg $(0.4x+0.4) - x^2$

$f'(x) = \cfrac{0.4}{\cos^2(0.4x+0.4)} - 2x$

$f''(x) = \cfrac{0 - 0.4 (\cos^2(0.4x+0.4))'}{\cos^4(0.4x+0.4)} -2 = \cfrac{-0.4 \cdot 2\cos(0.4x+0.4) \cdot (-\sin(0.4x+0.4)) \cdot 0.4}{\cos^4(0.4x+0.4)} - 2 = \cfrac{0.32 \sin(0.4x+0.4)}{\cos^3(0.4x+0.4)} - 2$

примітка: $f''(x)$ матиме такі ж вертикальні асимптоти як і $f(x)$ 

Побудуємо графік другої похідної


Як видно з графіка на проміжках (-0.5; -0.4) (1; 1.1) друга похідна функції збеігає знак. На проміжках (-5.1; -5) і (2.5; 2.6) вона також буде зберігати знак, бо кінці цих відрізків знаходяться по одну сторону від вертикальних асимптот $x = -4.92699$ і $x = 2.92699$ відповідно

```python
intervals = [(-5.1,-5), (-0.5, -0.4), (1, 1.1),  (2.5,  2.6)]

secant_2 = []

iter = 0 
step = 0

def second_derr_2(x):
    return (((0.32*math.sin(0.4*x+0.4))/((math.cos(0.4*x+0.4))**3)) -2)

for interval in intervals:
    a = interval[0]
    b = interval[1]
    
    if second_func(a)*second_derr_2(a) > 0:
        N = a
        x_k0 = b
    else:
        N = b
        x_k0 = a
        
    x_k1 = x_k0 + 2*eps
    
    while abs(x_k1 - x_k0) >= eps or abs(second_func(x_k1)) >= eps:
        
        if step != 0:
            x_k0 = x_k1
            
        x_k1 = x_k0 - ((second_func(x_k0))/(second_func(x_k0)-second_func(N)))*(x_k0 - N)
        
        step = 1
        iter = iter + 1
        
    print_(x_k1, second_func)
    secant_2.append(x_k1)
    
iterations.append(iter)
print(f"\nкількість ітерацій: {iter}")
        
        
```

```python
X = []
k = 0

intervals = [(-5.1,-5), (-0.5, -0.4), (1, 1.1),  (2.5,  2.6)]


print(" проміжок           п. наближення")

for interval in intervals:
    
    if second_func(interval[0])*second_derr_2(interval[0]) > 0:
        X.append(interval[0])
    else:
        X.append(interval[1])
        
    print(f"({interval[0]};{interval[1]})   \t\t {X[k]}")
    
    k = k + 1
    
```

```python
iter = 0 
step = 0
k = 0
newton_2 = []

def first_derr(x):
    return (((0.4)/((math.cos(0.4*x+0.4))**2)) - 2*x)

for interval in intervals:
    
    x_k0 = X[k]
    x_k1 = x_k0 + 2*eps
    
    while abs(x_k1 - x_k0) >= eps or abs(second_func(x_k1)) >= eps:
        
        if step != 0:
            x_k0 = x_k1
        
        x_k1 = x_k0 - (second_func(x_k0))/(first_derr(x_k0))

        step = 1
        iter = iter + 1
        
    k = k + 1
    step = 0
    print_(x_k1, second_func)
    newton_2.append(x_k1)
    
iterations.append(iter)
print(f"\nкількість ітерацій: {iter}")
```
